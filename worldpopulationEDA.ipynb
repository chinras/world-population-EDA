```python
#lets create a data frame for continents
data = df.drop(columns = ['CCA3', 'Capital', 'Country','Density (per km²)'])
continent = data.groupby('Continent').sum()
```

```python
continent['Density (per km²)'] = continent['2022 Population'] / continent['Area (km²)']
continent
```

```python
continent.head(6)
continent['Continents'] = continent.index


continent
#now lets work with visualization!
```

```python
sns.set_theme(style="whitegrid")
sns.barplot(x="Continents", y="2022 Population", data=continent)
#Asia has more population than all the continents combined
```

```python
rate = continent[['1970 Population','2022 Population','Continents']]



 
sns.lineplot(data=rate)
#we can see the rate of change from 1970 till 202
```
